```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.dtypes
```

```python
df = df.rename(columns={'Sex':'Gender'})
df.head()
```

```python
df.info()
```

```python
df['Age'].head(7)
```

```python
## converting datatype of an entire column

df['Age_int'] = df['Age'].astype(int)
df.head()
```

```python
## applying a function to an entire column

df['Fare_New'] = df['Fare'].apply(lambda x: x*2)
df.head()
```

```python
## data aggregation and grouping

grouped_mean = df.groupby('Pclass')['Fare'].mean()
grouped_mean
```

```python
## total aggregation based on multiple columns 
df.groupby(['Gender', 'Pclass'])['Survived'].sum()

# no of people that actually survived the Titanic based on their gender and Passenger Class
```

```python
## how to apply multiple aggregate fuinctions at once 

df.groupby('Survived')['Fare'].agg(['mean', 'sum', 'count'])
```

```python
## merging and joining 
df1 = pd.DataFrame({'Key':['A', 'B', 'C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A', 'B', 'D'], 'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## merging and joining dataframes
pd.merge(df1, df2, on="Key", how='inner')
```

```python
pd.merge(df1, df2, on='Key', how='outer')
```

```python
pd.merge(df1,df2,on='Key',how='left')
```

```python
pd.merge(df1,df2,on='Key',how='right')
```
